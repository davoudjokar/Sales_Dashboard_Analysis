```python
import pandas as pd
import io
import ipywidgets as widgets
from IPython.display import display

upload = widgets.FileUpload(
    accept='.csv',
    multiple=False
)

display(upload)

def handle_upload(change):
    if upload.value:
        # ‚úÖ ÿßÿµŸÑÿßÿ≠‚Äåÿ¥ÿØŸá ÿ®ÿ±ÿß€å ÿ™ÿßŸæŸÑ
        uploaded_file = upload.value[0]
        name = uploaded_file['name']
        content = uploaded_file['content']

        # ÿ™ÿ®ÿØ€åŸÑ ÿ®Ÿá DataFrame
        df = pd.read_csv(io.BytesIO(content))
        print(f"‚úÖ ŸÅÿß€åŸÑ '{name}' ÿ®ÿß ŸÖŸàŸÅŸÇ€åÿ™ ÿ®ÿßÿ±⁄Øÿ∞ÿßÿ±€å ÿ¥ÿØ.")
        display(df.head())

        # ÿ∞ÿÆ€åÿ±Ÿá global ÿ®ÿ±ÿß€å ÿßÿ≥ÿ™ŸÅÿßÿØŸá‚ÄåŸáÿß€å ÿ®ÿπÿØ€å
        globals()['df'] = df

upload.observe(handle_upload, names='value')
```

```python
# ŸÖ€åÿßŸÜ⁄Ø€åŸÜ Ÿáÿ≤€åŸÜŸá ŸÖÿßŸáÿßŸÜŸá ÿ®ÿ±ÿß€å Ÿáÿ± ŸÜŸàÿπ ŸÇÿ±ÿßÿ±ÿØÿßÿØ
df.groupby('Contract')['MonthlyCharges'].mean()

# ÿ™ÿπÿØÿßÿØ ŸÖÿ¥ÿ™ÿ±€å ÿ®ÿ±ÿßÿ≥ÿßÿ≥ ÿ¨ŸÜÿ≥€åÿ™
df['gender'].value_counts()

# ÿØÿ±ÿµÿØ ŸÖÿ¥ÿ™ÿ±€åÿßŸÜ€å ⁄©Ÿá churn ⁄©ÿ±ÿØŸÜ
df['Churn'].value_counts(normalize=True) * 100
```

```python
df['TotalCharges'] = pd.to_numeric(df['TotalCharges'], errors='coerce')
```

```python
df['AvgCostPerMonth'] = df['TotalCharges'] / (df['tenure'] + 1)
df[['customerID', 'TotalCharges', 'tenure', 'AvgCostPerMonth']].head()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

plt.figure(figsize=(8, 5))
sns.boxplot(data=df, x='Churn', y='AvgCostPerMonth')
plt.title('üí∞ Avg. Cost per Month by Churn')
plt.xlabel('Churn')
plt.ylabel('Average Cost per Month')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(df['AvgCostPerMonth'], bins=40, kde=True, color='skyblue')
plt.title('üìä Distribution of Avg. Cost per Month')
plt.xlabel('Avg. Cost per Month')
plt.ylabel('Count')
plt.grid(True)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.violinplot(data=df, x='gender', y='AvgCostPerMonth', hue='Churn', split=True)
plt.title('üéª Avg. Cost per Month by Gender & Churn')
plt.xlabel('Gender')
plt.ylabel('Average Cost per Month')
plt.grid(True)
plt.show()
```
